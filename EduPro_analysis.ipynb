```python
teachers=pd.read_excel("EduPro Online Platform.xlsx",sheet_name="Teachers")
courses=pd.read_excel("EduPro Online Platform.xlsx",sheet_name="Courses")
transactions=pd.read_excel("EduPro Online Platform.xlsx",sheet_name="Transactions")
print(teachers.head())
print(courses.head())
print(transactions.head())
```

```python
#merge
merged_df=pd.merge(transactions,teachers,on="TeacherID")
merged_df=pd.merge(merged_df,courses,on="CourseID")
print(merged_df.head())
```

```python
merged_df
```

```python
#teacher ratings
merged_df['TeacherRating'].describe()
```

```python
import matplotlib.pyplot as plt
merged_df['TeacherRating'].hist()
plt.title("Distribution of Teacher Ratings")
plt.xlabel("Rating")
plt.ylabel("count")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter(merged_df['YearsOfExperience'],merged_df['TeacherRating'])
plt.title("Experience vs Teacher Rating")
plt.xlabel("Years of Experience")
plt.ylabel("Teacher Rating")
plt.show()
```

```python
#TeacherRating and take average CourseRating
grouped = merged_df.groupby('TeacherRating')['CourseRating'].mean()

import matplotlib.pyplot as plt
grouped.plot(kind='line', marker='o')
plt.title("Average Course Rating vs Teacher Rating")
plt.xlabel("Teacher Rating")
plt.ylabel("Average Course Rating")
plt.show()
```

```python
expertise_performance = merged_df.groupby('Expertise')['CourseRating'].mean()
print(expertise_performance)
```

```python
expertise_performance.plot(kind='bar')
import matplotlib.pyplot as plt
plt.title("Course Rating by Expertise")
plt.xlabel("Expertise")
plt.ylabel("Average Course Rating")
plt.xticks(rotation=45)
plt.show()
```

```python
#experience and get average course rating
exp_course = merged_df.groupby('YearsOfExperience')['CourseRating'].mean()

import matplotlib.pyplot as plt
exp_course.plot(kind='line', marker='o')
plt.title("Experience vs Average Course Rating")
plt.xlabel("Years of Experience")
plt.ylabel("Average Course Rating")
plt.show()
```

```python
#performance category
def categorize_rating(rating):
    if rating >= 4.5:
        return "High"
    elif rating >= 3:
        return "Medium"
    else:
        return "Low"

#function
merged_df['PerformanceLevel'] = merged_df['TeacherRating'].apply(categorize_rating)
#counts
merged_df['PerformanceLevel'].value_counts()
```

```python
import matplotlib.pyplot as plt
merged_df['PerformanceLevel'].value_counts().plot(kind='bar')
plt.title("Instructor Performance Levels")
plt.xlabel("Performance Level")
plt.ylabel("Count")
plt.show()
```

```python
import matplotlib.pyplot as plt
category_performance = merged_df.groupby("CourseCategory")["CourseRating"].mean()
category_performance.plot(kind="bar")
plt.title("Average Course Rating by Category")
plt.xlabel("Course Category")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)
plt.show()
```

```python
level_performance = merged_df.groupby("CourseLevel")["CourseRating"].mean()
level_performance.plot(kind="bar")
plt.title("Course Rating by Level")
plt.xlabel("Course Level")
plt.ylabel("Average Rating")
plt.show()
```

```python
import pandas as pd

gender_level = pd.crosstab(merged_df["Gender"], merged_df["CourseLevel"])

gender_level.plot(kind="bar", stacked=True)
plt.title("Gender vs Course Level Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```
